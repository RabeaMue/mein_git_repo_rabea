## DataFrame filtern

```python
# Welche Dateien sind größer als 50 MB?
df_files["hasData_size_mb"] >= 50
```

```python
# Filtere DataFrame nach dieser Bedingung
df_files[df_files["hasData_size_mb"] >= 50]
```

```python
# Filtere Dataframe, aber zeig mir nur die genannten Spalten an
df_files[df_files["hasData_size_mb"] >= 50][["@id", "hasData_format", "hasData_size_mb"]]
```

```python
# Teilmenge als neuen Dataframe speichern
df_file_size_over_50 = df_files[df_files["hasData_size_mb"] >= 50][["@id", "hasData_format", "hasData_size_mb", "isDescribedBy_createdBy"]]
df_file_size_over_50.shape
```

```python
# Dataframe sortieren (und mit der Pandas-Doku arbeiten (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html))
df_file_size_over_50_sorted = df_file_size_over_50.sort_values(by=["hasData_size_mb"], ascending=False)
df_file_size_over_50_sorted
```

```python
# Werte einer Spalte ersetzen
df_file_size_over_50_sorted["isDescribedBy_createdBy"] = df_file_size_over_50_sorted["isDescribedBy_createdBy"].replace(288.0, "personA").replace(339.0, "personB").replace(22.0, "personC")
df_file_size_over_50_sorted
```

```python
# Dataframe als Datei speichern
df_file_size_over_50_sorted.to_csv("df_file_size_over_50_sorted.csv")
```

## Zwei Dataframes mergen

```python
# Datei für PMD-Daten herunterladen
# * PMD = PublikationsMetaDaten
url_data_pmd = "https://raw.githubusercontent.com/lschruff/library_carpentry_workshops/refs/heads/main/pandas_python_lesson/2025-12-01_frl_export_pmd.csv"
filename_pmd = "frl_export_pmd.csv"

urllib.request.urlretrieve(url_data_pmd, filename_pmd)
```

```python
# Zweiten Dataframe einlesen
df_pmd = pd.read_csv("frl_export_pmd.csv")
df_pmd.shape
```

```python
# Spaltenüberschriften des zweiten Dataframes
df_pmd.columns
```

```python
# Spaltenüberschriften des ersten Dataframes
df_files.columns
```

### Spaltenüberschriften fürs Mergen vorbereiten

```python
# Kopie beider Dataframes machen
df_files_copy = df_files.copy()
df_pmd_copy = df_pmd.copy()
```

```python
# Alle Spaltenüberschriften mit "_pmd" anreichern
for key in df_pmd_copy.columns:
    df_pmd_copy.rename(columns={key: f"{key}_pmd"}, inplace=True)
df_pmd_copy.columns
```

```python
# Alle Spaltenüberschriften außer "parentPid" mit "_file" anreichern
for key in df_files_copy.columns:
    if key == "parentPid":
        pass
    else:
        df_files_copy.rename(columns={key: f"{key}_file"}, inplace=True)
df_files_copy.columns
```

```python
# Schlüsselspalte umbenennen
df_pmd_copy.rename(columns={"@id_pmd": "parentPid"}, inplace=True)
```

```python
# Spaltenüberschriften des zweiten Dataframes erneut anzeigen
df_pmd_copy.columns
```

```python
# Dataframes mergen
df_frl_merged = df_files_copy.merge(df_pmd_copy, how="left", on="parentPid")
df_frl_merged.shape
```

```python
# Spaltenüberschriften anschauen
df_frl_merged.columns
```

```python
# Nur die Zeilen ausgeben, für die Dateigröße >= 50 MB
df_frl_merged[df_frl_merged["hasData_size_mb_file"] >= 50]
```

```python
df_frl_merged_size_over_50 = df_frl_merged[df_frl_merged["hasData_size_mb_file"] >= 50]
```

```python
df_frl_merged_size_over_50.to_csv("df_frl_merged_size_over_50.csv", index=False)
```
